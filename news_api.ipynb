```python
import requests
import pandas as pd
from translate import Translator

# 번역기 생성
translator = Translator(to_lang='ko')

# API 요청 설정
url = 'https://newsapi.org/v2/everything'
params = {
    'q': 'tennis',
    'sortBy': 'publishedAt',
    'from_param':'2023-05-15',
    'to':'2023-05-15',
#     'language' : 'ko',
    'apiKey': '420a0fa0297a44f384577be3d2e6799d'  # 여기에 실제 News API 키를 입력해야 합니다.
}

# API 요청 보내기
response = requests.get(url, params=params)
data = response.json()

# 응답 데이터를 데이터프레임으로 변환
articles = data['articles']
df = pd.DataFrame(articles)

# 필요한 컬럼만 선택
df = df[['title', 'source', 'publishedAt', 'description', 'url']]

# 'source' 컬럼을 문자열로 변환
df['source'] = df['source'].apply(lambda x: x['name'])

# 'title' 컬럼을 한글로 번역
# df['title_korean'] = df['title'].apply(lambda x: translator.translate(x))

df_head = df.head()
df_head

# 'title' 컬럼을 한글로 번역
df_head['title_korean'] = df_head['title'].apply(lambda x: translator.translate(x))
df_head
```

```python
import requests

# URI 설정
uri = '/v1/policies'
# '  # 실제 API URI로 대체해야 합니다.

# GET 요청 보내기
response = requests.get(uri)

# 응답 확인
if response.status_code == 200:
    # 응답이 성공적으로 수신된 경우, 응답 내용 출력
    print(response.json())
else:
    # 요청이 실패한 경우, 오류 메시지 출력
    print(f'Error: {response.status_code}')
```
